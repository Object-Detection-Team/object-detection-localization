```python
!pip install ultralytics
from ultralytics import YOLO
```

```python
model = YOLO('yolov8m.pt')

results = model.train(
    data='./drive/MyDrive/labeled-dataset/data.yaml',
    epochs=50,
    imgsz=640,
    batch=16,
    device=0
)
```

```python
import shutil
import os

# Define the destination folder in your Google Drive
drive_destination = '/content/drive/MyDrive/labeled-dataset/training_results'

# Create the directory if it does not exist
if not os.path.exists(drive_destination):
    os.makedirs(drive_destination)

# Copy the best performing model weights
shutil.copy('/content/runs/detect/train/weights/best.pt', f'{drive_destination}/fruit_model_best.pt')

# Copy the training result visualizations (Loss, mAP curves, etc.)
shutil.copy('/content/runs/detect/train/results.png', f'{drive_destination}/results.png')
shutil.copy('/content/runs/detect/train/confusion_matrix.png', f'{drive_destination}/confusion_matrix.png')

print(f"Model and training reports have been successfully saved to: {drive_destination}")
```
